# Exercise 6 - geopandas

```python
import geopandas as gpd
```

```python
filename = '../datasets/wijkenbuurten_2024_v1.gpkg'
```

```python
gpd.list_layers(filename)
```

```python
cbs = gpd.read_file(filename, layer='gemeenten')
```

```python
cbs.info(verbose=True)
```

```python
cbs.plot()
```

```python
gemeenten = cbs[cbs['water']=='NEE'].copy()
```

```python
gemeenten.plot()
```

```python
gemeenten.plot('bevolkingsdichtheid_inwoners_per_km2', figsize=(10, 10));
```

```python
gemeenten['man_vrouw'] = gemeenten['mannen'] / gemeenten['vrouwen']
```

```python
gemeenten.plot('man_vrouw', figsize=(8, 8), legend=True)
```

```python
gemeenten.nlargest(10, 'man_vrouw')[['gemeentenaam', 'man_vrouw']]
```

```python
gemeenten.nsmallest(10, 'man_vrouw')[['gemeentenaam', 'man_vrouw']]
```
