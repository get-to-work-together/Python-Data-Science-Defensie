# Demo functies

### Definitie

```python
def doe_iets():
    print('Hoi')
    print('Gaat het goed?')
    print('Doei.')
```

### Aanroepen

```python
doe_iets()
```

### Met argumenten

```python
def doe_iets(naam, n = 1):
    for i in range(n):
        print('Hoi', naam)
    print('Gaat het goed?')
    print('Doei.')
```

```python
doe_iets('Peter', 4)
```

### Return value

```python
def bereken_oppervlakte(lengte, breedte):
    return lengte * breedte
```

```python
bereken_oppervlakte(3, 4)
```

```python
bereken_oppervlakte(breedte = 4, lengte = 3)
```

```python
def count_a(woord):
    return woord.count('a')
```

```python
woorden = 'de grote bruine hond slaapt in zijn hok'.split()
woorden
```

```python
sorted(woorden)
```

```python
sorted(woorden, key = len)
```

```python
sorted(woorden, key = count_a)
```

```python
sorted(woorden, key = lambda woord: woord.count('e'))
```
