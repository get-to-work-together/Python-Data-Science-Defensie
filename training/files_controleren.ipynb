```python
def get_all_files(pad):
    all_files = []
    for root, dirs, files in os.walk(pad, topdown=True, followlinks=False):
        dirs[:] = [d for d in dirs if not d.startswith(".")]
        files[:] = [f for f in files if not f.startswith(".")]
        files[:] = [f for f in files if not f.startswith("~")]
    
        # print(f"Directory path: {root}")
        # print(f"Directory Names: {dirs}")
        # print(f"Files Names: {files}")
        # print(80*'-')
    
        for file in files:
            path = '/'.join([root, file])
                            
            info = Path(path)
            user = info.owner()
            group = info.group()
            
            # Get file stats
            stat = os.stat(path)
            
            # File size in bytes
            size = stat.st_size
            
            # Owner UID
            uid = stat.st_uid
            
            # Creation time (platform dependent)
            ctime = stat.st_ctime  # On Unix: metadata change time, on Windows: creation time
            
            # Modification time
            mtime = stat.st_mtime
            
            # Convert timestamps to human-readable
            ctime_str = time.ctime(ctime)
            mtime_str = time.ctime(mtime)
    
            all_files.append({
                "File": file,
                "Directory": root,
                "Size": size,
                "Owner": uid,
                "User": user,
                "Group": group,
                "Created": ctime,
                "Modified": mtime,          
            })
            
            # print(f"File: {file}")
            # print(f"Directory: {root}")
            # print(f"Size: {size} bytes")
            # print(f"Owner: {uid}")
            # print(f"User: {user}")
            # print(f"Group: {group}")
            # print(f"Created: {ctime_str}")
            # print(f"Modified: {mtime_str}")
            # print(80 * '-')
    
    return all_files
```

```python
all_files = get_all_files(pad)
```

```python
df = pd.DataFrame(all_files)
```

```python
df.to_excel('all_files.xlsx')
```

```python
subset = ['Size']
duplicates = df.duplicated(subset=subset, keep=False)
df.loc[duplicates, 'duplicated'] = df[subset].apply(tuple, axis=1)
```

```python
regex = r'(\d{8})_(\w{2})_.*'
df[['filename_dd', 'filename_classificatie']] = df['File'].str.extract(regex)
```
