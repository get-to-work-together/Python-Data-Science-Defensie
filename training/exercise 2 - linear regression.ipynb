# Exercise 2 - Lineaire Regressie

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Genereer data

```python
a = 2
b = 5

def onderliggend_model(x):
    return a * x + b
```

```python
n = 100

x_min = 0
x_max = 10

x = np.random.uniform(x_min, x_max, n)

y = onderliggend_model(x)
```

```python
spread = 3

ruis = np.random.randn(n) * spread

y = y + ruis
```

```python
plt.scatter(x, y)
plt.show()
```
