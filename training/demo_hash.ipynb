# Demo HASH

```python
wachtwoord = 'Welkom!1234'
```

```python
hash_wachtwoord = hash(wachtwoord)
hash_wachtwoord
```

```python
invoer = input('Geef wachtwoord: ')
if hash(invoer) == hash_wachtwoord:
    print('Dat is goed')
else:
    print('Verkeerde wachtwoord')
```

```python
import hashlib
```

```python
hash_wachtwoord = hashlib.sha512(wachtwoord.encode()).hexdigest()
hash_wachtwoord
```

```python
invoer = input('Geef wachtwoord: ')
hash_invoer = hashlib.sha512(invoer.encode()).hexdigest()
if hash_invoer == hash_wachtwoord:
    print('Dat is goed')
else:
    print('Verkeerde wachtwoord')
```
