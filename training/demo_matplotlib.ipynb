### EDA - Exploratory Data Analysis

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

ax = axs[0]
ax.scatter(df['sepal_length'], df['sepal_width'], label='sepal', color='red', marker='+', s=20)
ax.legend()
ax.grid()
ax.set_title('sepal')

ax = axs[1]
ax.scatter(df['petal_length'], df['petal_width'], label='petal', color='orange', marker='d', s=200, alpha=0.4)
ax.legend()
ax.grid()
ax.set_title('petal')

plt.show()
```

```python
x = np.linspace(0, 10, 40)

y1 = np.sin(x)
y2 = np.sin(2 * x) + 2
y3 = np.sin(3 * x) + 4
y4 = np.sin(4 * x) + 6
```

```python
fig, axs = plt.subplots(4, 1, sharex=True)

axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, y3)

axs[3].plot(x, y4)

for ax in axs:
    ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(sharex=True)


ax.plot(x, y1, linewidth=3, linestyle='dotted')
ax.plot(x, y2, marker=TextPath((0, 0), 'ðŸ˜€'), markersize=40)
ax.plot(x, y3)
ax.plot(x, y4)

ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(sharex=True)

ax.scatter(y1, y2, c='C2', s=200, alpha=0.6, edgecolor='k')

ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots()

bins = np.linspace(0, 8, 25)
ax.hist(df['sepal_length'], bins=bins, edgecolor='darkblue', zorder=10, alpha=0.7)
ax.hist(df['petal_length'], bins=bins, edgecolor='darkblue', zorder=10, alpha=0.7)

ax.grid()
ax.set_title('sepal_length')
ax.set_xlabel('mm')

plt.show()
```
