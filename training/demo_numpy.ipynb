```python
data = np.random.randn(1000)
```

```python
data
```

```python
np.mean(data)
```

```python
np.std(data)
```

```python
data[50] = 10
```

```python
data[40:60]
```

```python
(data < np.mean(data) - 3 * np.std(data)) | (data > np.mean(data) + 3 * np.std(data))
```

```python
data[(data < np.mean(data) - 3 * np.std(data)) | (data > np.mean(data) + 3 * np.std(data))]
```

```python
median = np.median(data)

q1 = np.quantile(data, q=0.25)
q3 = np.quantile(data, q=0.75)

print(q1, median, q3)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print(lower, upper)
```

```python
plt.boxplot(data)
plt.show()
```
