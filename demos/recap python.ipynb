### Comprehensions

* Met een list comprehension schrijfwijze kan op een compacte wijze een lijst worden gemaakt van een al bestaande lijst.
* Een comprehension lijkt op een for-loop binnen de haakjes van een datastructuur.
* Er kan ook een criterium worden toegevoegd. Alleen elementen die aan de criterium voldoen komen in de nieuwe lijst.
* Er zijn ook comprehensions voor een set en een dict

```python
getallen = [2, 4, 7, 8, 9]
print(getallen)

kwardraten = [getal ** 2 for getal in getallen]
print(kwardraten)

kwardraten_van_de_even_getallen = [getal ** 2 for getal in getallen if getal % 2 == 0]
print(kwardraten_van_de_even_getallen)
```

```python
keys = ['nl', 'b', 'f']
values = ['+31', '+32', '+33']
d = {k:v for k, v in zip(keys, values)}

print(d)
```

### Argumenten en return waarde
* In de definitie van een functie kunnen argumenten worden opgegeven die binnen de functie kunnen worden gebruikt.
* Een functie kan een waarde teruggeven met de **return** keyword.
* Een functie wordt uitgevoerd door de naam van de functie te geven gevolgd door ronde haken met daarin eventuele argumenten.
* De argumenten kunnen op basis van volgorde worden meegegeven (positionele argumenten).
* Argumenten kunnen ook met de naam van het argument worden opgegeven (keyword arguments). Dan maakt de volgorde niet uit.

‚©• https://www.w3schools.com/python/python_arguments.asp

```python
def bereken_volume(lengte, breedte, hoogte):
    return lengte * breedte * hoogte

volume = bereken_volume(10, 20, 30)
print(volume)

volume = bereken_volume(hoogte=30, lengte=10, breedte=20)
print(volume)
```

### Default waarden

* Argumenten kunnen default waarden hebben.
* Als een argument in de aanroep niet wordt meegegeven dan wordt de default waarde gebruikt
* Hierdoor is het mogelijk om alleen die argumenten mee te geven die een waarde moeten krijgen die anders is dan de default waarde

```python
def book_flight(from_airport, to_airport, num_adults=1, num_children=0):
    print(f"\nFlight booked from {from_airport} to {to_airport}")
    print(f"Number of adults: {num_adults}")
    print(f"Number of children: {num_children}")
    
book_flight(from_airport="BRS", to_airport="VER", num_adults=2, num_children=2)
book_flight("LHR", "CDG", num_children=2)
book_flight(num_children=3, from_airport="LGW", to_airport="NCE")
```

### Variadic arguments

* Een wisselend aantal positionele argumenten kunnen worden opgegeven met een * voor het argument.
* De meegegeven argumenten komen dan in een tuple.
* Ongespecificeerde keyword argumenten kunnen worden gebruikt met ** voor het argument.
* De meegegeven keyword argumenten komen dan in een dict

‚©• https://www.w3schools.com/python/python_args_kwargs.asp

```python
def toon_argumenten(positionele_argument, *args, keyword_argument, **kwargs):
    print(positionele_argument)
    print(args)
    print(keyword_argument)
    print(kwargs)

toon_argumenten('Appel', 'Bannan', 'Citroen', 
                keyword_argument='Wijn' , 
                keyword_argument2='Bier', 
                keyword_argument3='Spa')
```

### Functies als een waarde

* Functies kunnen net als getallen, strings en datastructuren als een waarde worden toegewezen aan variabelen.
* Functies kunnen als een waarde worden opgenomen in datastructuren.
* Functies kunnen als argument worden meegegeven aan een andere functie. Vaak wordt dan een **lambda** functie gebruikt.

```python
woorden = 'de kat krabt de krullen van de trap'.split()

print(woorden)
print(sorted(woorden))                                 # alfabetisch gesorteerd
print(sorted(woorden, key=len))                        # gesorteerd op lengte
print(sorted(woorden, key=lambda w: w.count('a')))     # gesorteerd het aantal a's
```

### Het lezen van een bestand

```python
filename = 'data.csv'

with open(filename, mode='w') as f:
    content = """\
id,naam,woonplaats
1,Guido,Amsterdam
2,Gerard,Utrecht
3,Sandra,Rotterdam
4,Jantje,Amsterdam"""
    f.write(content)
```
